## 3. Feature Engineering

Com base no conjunto de dados, alguns novos recursos (features) podem ser criados de forma a melhorar o desempenho e a capacidade preditiva do modelo.<br>

Um importante conceito no mundo da segmentação de clientes é a Análise RFM que consiste em mensuar três fatores com relação ao comportamento do cliente:
- Recency (ou Recência): mede o quão recente um cliente fez uma compra ou interagiu com a empresa.
- Frequency (ou Frequência): mede a frequência com que um cliente faz compras ou interage com a empresa.
- Monetary (ou Monetário): mede o valor monetário total das compras ou transações de um cliente com a empresa.

Iremos então criar novas colunas no dataframe que quantificam esses valores

### 3.1. Recency
Vamos calcular o número de dias desde a útima compra pra cada cliente

```python
# Converter InvoiceDate para datetime
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])

# Encontrar a data da compra mais recente para cada cliente
clientes = df.groupby('CustomerID')['InvoiceDate'].max().reset_index()

# Encontrar a data mais recente no dataframe
data_mais_recente = df['InvoiceDate'].max()

# Calcular a diferença de dias desde a última compra para cada cliente
clientes['dias_desde_ultima_compra'] = (data_mais_recente - clientes['InvoiceDate']).dt.days
clientes.drop('InvoiceDate', axis=1, inplace=True)

clientes.head()
```

### 3.2. Frequency
Neste caso, iremos calcular o número de transações por cliente

```python
# Período de ados analisados
periodo = df['InvoiceDate'].max() - df['InvoiceDate'].min()

# Número de transações por cliente
total_transacoes = df.groupby('CustomerID')['InvoiceNo'].nunique().reset_index()

# Juntando com o dataframe 'clientes' criado anteriormente
clientes = pd.merge(clientes, total_transacoes, on='CustomerID')

# # Renomenando coluna
clientes.rename(columns={'InvoiceNo':'num_transacoes'}, inplace=True)

clientes.head()
```

### 3.3. Monetary
Aqui, iremos calcular o total gasto por cliente

```python
# Criando nova coluna com o valor total da transação
df['valor_transacao'] = df['Quantity']*df['UnitPrice']

# Calculando o total por cliente
total_gasto = df.groupby('CustomerID')['valor_transacao'].sum().reset_index()

# Juntando com o dataframe 'clientes' criado anteriormente
clientes = pd.merge(clientes, total_gasto, on='CustomerID')

# Renomenando coluna
clientes.rename(columns={'valor_transacao':'total_gasto'}, inplace=True)

clientes.head()
```

- Além das métricas citadas anteriormente, vamos criar mais algumas que podem melhorar o desempenho do algoritmo.

### 3.4. Gasto médio mensal
Além do cálculo do gasto médio mensal, também é relevante analisar o desvio padrão dos gastos mensais. O desvio padrão oferece uma visão sobre a variação nos gastos do cliente: um valor elevado pode indicar sazonalidade no comportamento de compras, enquanto um valor baixo sugere que o cliente mantém hábitos de compra mais consistentes ao longo do tempo.

```python
# Criando novas colunas com o mês da transação
df['mes'] = df['InvoiceDate'].dt.month

# Calculando o gasto médio mensal e desvio padrão por cliente
gasto_mensal = df.groupby(['CustomerID','mes'])['valor_transacao'].sum().reset_index()
gasto_mensal_medio_std = gasto_mensal.groupby('CustomerID')['valor_transacao'].agg(['mean','std']).reset_index()

# Quando o cliente possui transação em apenas 1 mês, o resultado de std será NaN, pois para calcular o desvio padrão, é necessário pelo menos 2 valores
# Substituindo os valores de NaN por 0
gasto_mensal_medio_std['std'] = gasto_mensal_medio_std['std'].fillna(0)

# Juntando com o dataframe 'clientes' criado anteriormente
clientes = pd.merge(clientes, gasto_mensal_medio_std, on='CustomerID')

# Renomenando coluna
clientes.rename(columns={'mean':'media_gasto_mensal', 'std':'std_gasto_mensal'}, inplace=True)

clientes.head()
```

### 3.5. Volume de itens adquiridos
Esta métrica pode ser útil no processo de agrupamento de clientes, pois ao mostrar a quantidade total de itens adquiridos por cliente, permite identificar os mais fiéis e com alto valor para a loja.

```python
# Contagem de produtos adquiridos por cliente
produtos = df.groupby('CustomerID')['Quantity'].sum().reset_index()

# Juntando com o dataframe 'clientes' criado anteriormente
clientes = pd.merge(clientes, produtos, on='CustomerID')

# Renomenando coluna
clientes.rename(columns={'Quantity':'volume_itens_adquiridos'}, inplace=True)

clientes.head()
```

## 4. Outliers
Nesse passo, iremos utilizar uma técnica de aprendizado de máquina chamado Isolation Forest para detecção de outliers. Algumas das principais vantagens desse algoritmo, são:

- Não requer rótulos para treinamento, funcionando bem com dados não supervisionados.
- Efetivo para conjuntos de dados de alta dimensão.

Isso faz com que a escolha desse algoritmo seja bastante plausível, tendo em vista que se encaixa bem no caso em estudo.

```python
# Criar e treinar o modelo Isolation Forest
taxa_contaminacao = 0.05
modelo_IF = IsolationForest(contamination = taxa_contaminacao, random_state=11)  # 'contamination' define a proporção esperada de anomalias
modelo_IF.fit(clientes)

# Crindo array com as predições
predicoes = modelo_IF.predict(clientes)

# Adicionando a array 'predicoes' no dataframe de clientes
clientes['outlier'] = predicoes

# Substituindo os valores 1 para outlier e -1 para não-outlier
clientes['outlier'] = clientes['outlier'].apply(lambda x: 1 if x==-1 else 0)

clientes['outlier'].value_counts()
```

- Como a taxa de contaminação foi definida como 5%, o algoritmo retornou 219 clientes como outliers. Vamos então remover os outliers do dataframe

```python
# Removendo os outliers
clientes_sem_outlier = clientes[clientes['outlier']==0]
outliers = clientes[clientes['outlier']==1]

# Retirando a coluna outlier do dataset
clientes_total = clientes_sem_outlier.drop('outlier', axis=1)
```

## 6. Normalizando os valores
A normalização dos dados é crucial em muitos modelos de Machine Learning. Isso é feito para garantir que todas as variáveis estejam na mesma escala, especialmente no contexto de algoritmos baseados em distância, como K-means. 

Quando os dados não estão em uma escala semelhante, os recursos com valores maiores podem influenciar desproporcionalmente o resultado do agrupamento, levando potencialmente a agrupamentos incorretos.

Contudo, alguns recursos não requerem normalização, como por exemplo CustomerID e dia_favorito. CustomerID por ser apenas um identificador e dia_favorito por ser uma variável categória e representar apenas números inteiros de 1 a 7.

```python
# criando o normalizador
scaler = StandardScaler()

# Features que não necessitam de normalização
nao_normalizar = ['CustomerID']

# Selecionando apenas colunas para normalizar
aux = clientes_final.drop(nao_normalizar, axis=1)
normalizar = aux.columns

# Criando cópia do dataframe
clientes_normal = clientes_final.copy()

# ajustando o conjunto de dados no normalizador apenas com as colunas a normalizar
clientes_normal[normalizar] = scaler.fit_transform(clientes_final[normalizar])

clientes_normal.head()
```

## 7. Análise de correlação e redução de dimensionalidade
Agora, vamos verificar se há alta correlação entre as colunas do dataset

```python
# Criando matriz de correlação
plt.figure(figsize=(15, 10))
sns.heatmap(clientes_normal.corr(), annot=True, cmap='vlag_r', center=0)
plt.show()
```

**Sendo assim, é recomendável que seja realizada uma redução de dimensionalidade, pois a presença de multicolinearidade, onde os recursos são altamente correlacionados, pode afetar potencialmente o processo de agrupamento, não permitindo que o modelo aprenda corretamente os padrões. Isto pode levar a clusters que não são bem separados e significativos.**

Dentre os diversos métodos de redução de dimensionalidade, usaremos o PCA (Principal Component Analysis). O PCA é um método linear que remove a correlação entre as variáveis, transformando-as em um conjunto de componentes principais ortogonais (não correlacionados), mantendo a maior parte da variância dos dados originais.

```python
# Colocando CustomerID como índice
clientes_normal.set_index('CustomerID', inplace=True)
```

```python
# Aplicar PCA
pca = PCA().fit(clientes_normal)

# Calcular a Soma Acumulada da Variância Explicada
explained_variance = pca.explained_variance_ratio_
explained_variance_acumulada = np.cumsum(explained_variance)

# Plotar a Variância Explicada vs Número de Componentes
plt.figure(figsize=(8, 6))
plt.plot(explained_variance_acumulada, marker='o', linestyle='--')
plt.xlabel('Número de Componentes')
plt.ylabel('Variância Explicada Acumulada')
plt.title('Soma Acumulada da Variância Explicada pelo PCA')
plt.grid(True)
plt.show()

# Escolher o número de componentes (se o aumento na Variância Explicada Acumulada for menor que 0.1)
for i in range(len(explained_variance_acumulada)-1):
    diferenca = explained_variance_acumulada[i+1] - explained_variance_acumulada[i]
    if diferenca < 0.1:
        num_componentes = i+1
        break

print(f'Número de componentes a ser usado: {num_componentes}')
```

```python
# Transformar os dados para o número de componentes selecionado
pca = PCA(n_components=num_componentes)
dados_pca = pca.fit_transform(clientes_normal)

# Criando o dataframe com os dados reduzidos
clientes_pca = pd.DataFrame(dados_pca, columns=[['C'+str(i+1) for i in range(pca.n_components_)]], index=clientes_normal.index)

clientes_pca.head()
```

## 8. Clusterização com K-Means
Agora que já foram realizados todos os passos de limpeza e preparação dos dados, podemos prosseguir para o processo de criação do modelo de clusterização utilizando o K-Means. Alguns hyperparâmetros serão configurados para melhorar o modelo como:
- Várias Inicializações (n_init): serve para estabilizar os resultados do K-Means especificando o número de vezes que o algoritmo será executado com diferentes centroides aleatórios. A solução com o menor erro de clusters será escolhida.
- K-Means++: Um método de inicialização que melhora a escolha inicial dos centroides para aumentar a probabilidade de convergência para uma solução globalmente ótima.

Durante a utilização do K-Means, é necessário informar o número de clusters ou agrupamentos que o algoritmo deve considerar. Para fazer isso utilizaremos:
- Método do cotovelo (Elbow Method)
- Silhouette Score

### Utilizando Elbow Method
Este método representa a soma das distâncias ao quadrado de cada ponto de dados ao centroide do seu respectivo cluster. Em outras palavras, mede o quão compactos os clusters são. O ponto onde a redução do distortion score começa a se estabilizar (ponto de cotovelo) indica o número ideal de clusters.

```python
# Criando um modelo KMeans
kmeans = KMeans(init='k-means++', n_init=10, random_state=0)

# Criando o visualizador com o modelo 
elbow = KElbowVisualizer(kmeans, k=10, timings=False)

# Treinando o visualizador com os dados
elbow.fit(clientes_pca)

# Exibindo o visualizador
elbow.show()
```

### Utilizando Silhouette Score
Mede a similaridade entre pontos de um mesmo cluster em comparação com os pontos dos outros clusters. Quanto mais alto o score, melhor será o número de clusters.

```python
silhouette_scores = []
k_values = range(3, 11)

for k in k_values:
    kmeans = KMeans(n_clusters=k, init='k-means++', n_init=10, random_state=0)
    clusters = kmeans.fit_predict(clientes_pca)
    score = silhouette_score(clientes_pca, clusters)
    silhouette_scores.append(score)

# Plotando o Silhouette Score
plt.figure(figsize=(8, 6))
plt.plot(k_values, silhouette_scores, marker='o', linestyle='--')
plt.xlabel('Número de Clusters')
plt.ylabel('Silhouette Score')
plt.title('Silhouette Score para Diferentes Números de Clusters')
plt.grid(True)
plt.show()

# Melhor número de clusters
melhor_k = k_values[np.argmax(silhouette_scores)]
print(f'O melhor número de clusters é: {melhor_k}')
```

- **Ambos os métodos retornaram o valor de k=5 como sendo o melhor número de agrupamentos.No entanto, ao analisar o gráfico com os valores de Silhouette Score, percebe-se que há uma diferença mínima entre os scores para 3, 4 e 5 clusters. Dessa forma, optar por k=3 não impactará negativamente as predições do algoritmo e pode, inclusive, ser mais vantajoso em alguns casos, pois tende a generalizar melhor.**

- **Escolher k=5 pode introduzir uma leve confusão entre os clusters. Por outro lado, a escolha de k=3 nos proporciona clusters mais equilibrados e bem definidos, tornando a solução de clusterização mais robusta e confiável. Vale destacar que a escolha do valor de k muitas vezes é guiada por considerações práticas ou comerciais, além de apenas métricas numéricas.**

```python
# Aplicando KMeans com k=3
kmeans = KMeans(n_clusters=3, init='k-means++', n_init=10, random_state=11)
clusters = kmeans.fit_predict(clientes_pca)
```

- **Aqui temos um problema que pode muitas das vezes passar despercebido. O algoritmo K-Means, devido à sua inicialização aleatória dos centroides, pode atribuir rótulos diferentes aos clusters em cada execução, mesmo que os clusters sejam os mesmos em termos de composição. Por exemplo, o cluster 0 em uma execução pode ser o cluster 2 em outra.**
- **Para garantir a consistência dos rótulos entre diferentes execuções, realizamos um mapeamento adicional dos rótulos baseado na frequência de amostras em cada cluster.**

```python
# Contamos quantos pontos de dados existem em cada cluster usando
contagem_clusters = Counter(kmeans.labels_)

# Criamos um dicionário que remapeia os rótulos dos clusters originais para novos rótulos, ordenados pela frequência.
rotulos_remap = {rotulo: novo_rotulo for novo_rotulo, (rotulo, _) in 
                 enumerate(contagem_clusters.most_common())}

# Usamos o mapeamento criado para substituir os rótulos originais pelos novos rótulos consistentes:
novos_rotulos = np.array([rotulos_remap[rotulo] for rotulo in kmeans.labels_])

# Adição dos Rótulos aos DataFrames
clientes_final['cluster'] = novos_rotulos
clientes_pca['cluster'] = novos_rotulos

clientes_final.head()
```

Vamos utilizar duas métricas para avaliar o modelo:
- Calinski Harabasz Score: Mede a separação entre clusters e a coesão dentro dos clusters. Valores mais altos são melhores
- Davies Bouldin Score: Mede a similaridade entre clusters. Valores mais baixos são melhores
- Silhouette Score: Mede a similaridade entre pontos de um mesmo cluster em comparação com os pontos dos outros clusters. Valores mais altos são melhores.

```python
X = clientes_pca.drop('cluster', axis=1)

chs = calinski_harabasz_score(X, clusters)
dbs = davies_bouldin_score(X, clusters)
ss = silhouette_score(X, clusters)

print(f'Calinski Harabasz Score: {chs:.4f}')
print(f'Davies Bouldin Score: {dbs:.4f}')
print(f'Silhouette Score: {ss:.4f}')
```

### 11. Sistema de recomendação
Utilizar um sistema de recomendação é uma estratégia de marketing bastante útil, pois proporciona uma melhor experiência de compra por parte dos clientes, além de aumentar as chances de conversão, oferecendo um produto que o usuário provavelmente possui interesse.

O algoritmo criará recomendações personalizadas para o dataset clientes_amostra criado anteriormente, sugerindo os três produtos mais vendidos em seu cluster que ainda não foram adquiridos pelo cliente.

```python
# Refazendo os mesmos passos, mas agora para os dados da amostra que ainda não foram clusterizados

# Selecionando apenas colunas para normalizar
normalizar = clientes_amostra.drop(nao_normalizar, axis=1).columns

# Criando cópia do dataframe
clientes_n = clientes_amostra.copy()

# ajustando o conjunto de dados no normalizador apenas com as colunas a normalizar
clientes_n[normalizar] = scaler.fit_transform(clientes_amostra[normalizar])

clientes_n.set_index('CustomerID', inplace=True)

# Reduzindo a dimensionalidade
pca_amostra = pca.fit_transform(clientes_n)

# Criando o dataframe com os dados reduzidos
pca_amostra = pd.DataFrame(pca_amostra, columns=[['C'+str(i+1) for i in range(pca.n_components_)]], index=clientes_amostra.index)

# Predizer os clusters para os clientes da amostra
clusters_amostra = kmeans.predict(pca_amostra)

# Contamos quantos pontos de dados existem em cada cluster usando
contagem_clusters_amostra = Counter(clusters_amostra)

# Criamos um dicionário que remapeia os rótulos dos clusters originais para novos rótulos, ordenados pela frequência.
rotulos_remap_amostra = {rotulo: novo_rotulo for novo_rotulo, (rotulo, _) in enumerate(contagem_clusters_amostra.most_common())}

# Usamos o mapeamento criado para substituir os rótulos originais pelos novos rótulos consistentes:
novos_rotulos_amostra = np.array([rotulos_remap_amostra[rotulo] for rotulo in clusters_amostra])

# Adição dos Rótulos ao DataFrame
clientes_amostra['cluster'] = novos_rotulos_amostra

clientes_amostra.head()
```

```python
# Criando cópia do dataframe original
df_aux = df.copy()

# Juntando clientes_amostra e clientes_final as informações de cluster
clientes_total = pd.concat([clientes_amostra, clientes_final])

# Extrair o CustomerID dos outliers e remover suas transações do dataframe principal
id_outliers = outliers['CustomerID']
df_aux = df_aux[~df_aux['CustomerID'].isin(id_outliers)]

# Adicionando informação do cluster para cada transação no dataframe principal
df_aux = df_aux.merge(clientes_total[['CustomerID', 'cluster']], on='CustomerID', how='inner')

# Identificando os 10 produtos mais vendidos em cada cluster com base na quantidade total vendida
top_10_produtos = df_aux.groupby(['cluster', 'StockCode', 'Description'])['Quantity'].sum().reset_index()
top_10_produtos = top_10_produtos.sort_values(by=['cluster', 'Quantity'], ascending=[True, False])
top_10_produtos_por_cluster = top_10_produtos.groupby('cluster').head(10)

# Criando um registro de produtos adquiridos por cada cliente em cada cluster
compras_clientes = df_aux.groupby(['CustomerID', 'cluster', 'StockCode'])['Quantity'].sum().reset_index()

# Recomendações para cada cliente da amostra em cada cluster
recomendacoes = []
for cluster in top_10_produtos_por_cluster['cluster'].unique():
    produtos = top_10_produtos_por_cluster[top_10_produtos_por_cluster['cluster'] == cluster]
    clientes = clientes_amostra[clientes_amostra['cluster'] == cluster]['CustomerID']
    
    for cliente in clientes:
        # Identificar produtos já comprados pelo cliente
        produtos_comprados = compras_clientes[(compras_clientes['CustomerID'] == cliente) & (compras_clientes['cluster'] == cluster)]['StockCode'].tolist()
            
        # Filtrar os produtos top 10 do cluster atual que o cliente ainda não comprou
        top_produtos_cluster = produtos[~produtos['StockCode'].isin(produtos_comprados)]
        
        # Selecionar os top 3 produtos do cluster atual
        top_3_produtos = top_produtos_cluster.head(3)
        
        # Adicionando as recomendações à lista
        recomendacoes.append([cliente, cluster] + top_3_produtos[['StockCode', 'Description']].values.flatten().tolist())

# Criando o dataframe com as recomendações
df_recomendacoes = pd.DataFrame(data=recomendacoes, columns=['CustomerID', 'cluster', 
                                                             'Rec1_StockCode', 'Rec1_Description', 
                                                             'Rec2_StockCode', 'Rec2_Description', 
                                                             'Rec3_StockCode', 'Rec3_Description'])
```
